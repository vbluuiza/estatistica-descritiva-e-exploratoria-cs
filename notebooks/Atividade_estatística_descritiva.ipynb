## 3. Análise de Quartis e Outliers

Os quartis dividem um dataset ordenado em quatro partes iguais (Q1: 25%, Q2: 50% - mediana, Q3: 75%). Eles são cruciais para entender a dispersão e identificar possíveis outliers. Um outlier é um ponto de dado que se desvia significativamente de outras observações. Podemos identificá-los usando o Intervalo Interquartil (IQR), que é a diferença entre Q3 e Q1, e estabelecendo limites inferior e superior.

### Exercício 3.1: Cálculo Manual de Quartis e Identificação de Outliers

Para a coluna 'total_bill', calcule manualmente (sem usar `describe()` diretamente para os quartis, mas pode usar `quantile()`) os valores de Q1, Q2 (mediana), Q3 e o Intervalo Interquartil (IQR). Em seguida, calcule o Limite Inferior e Superior para identificação de outliers (Limite Inferior = Q1 - 1.5 * IQR; Limite Superior = Q3 + 1.5 * IQR). Por fim, mostre quantos outliers existem na coluna 'total_bill'.

```python
df = sns.load_dataset('tips')
print("Dataset 'Tips' carregado com sucesso!")

q1 = df['total_bill'].quantile(0.25)
q2 = df['total_bill'].median()
q3 = df['total_bill'].quantile(0.75)
iqr = q3 - q1

limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

outliers = df[(df['total_bill'] < limite_inferior) | (df['total_bill'] > limite_superior)]

print(f"\n--- Resultados do Exercício 3.1 ---")
print(f"Q1: {q1:.2f} | Q3: {q3:.2f} | IQR: {iqr:.2f}")
print(f"Limite Superior: {limite_superior:.2f}")
print(f"Quantidade de Outliers encontrados: {len(outliers)}")
print("\nExemplos de registros considerados outliers:")
print(outliers.head())
```

## 4. Assimetria (Skewness) e Curtose

A assimetria (skewness) mede o grau de inclinação de uma distribuição de probabilidade. Uma distribuição simétrica tem skewness zero. Valores positivos indicam uma cauda direita mais longa, enquanto valores negativos indicam uma cauda esquerda mais longa.

A curtose (kurtosis) mede o 'achatamento' ou 'pico' de uma distribuição. Uma curtose alta significa mais valores nas caudas e um pico mais nítido (leptocúrtica), enquanto uma curtose baixa indica menos valores nas caudas e um pico mais plano (platocúrtica). Uma distribuição normal tem curtose próxima de 3 (ou 0, se for calculada a curtose em excesso).

### Exercício 4.1: Cálculo e Interpretação de Assimetria e Curtose

Calcule a assimetria (skewness) e a curtose para as colunas numéricas ('total_bill', 'tip', 'size').

```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')

colunas_alvo = ['total_bill', 'tip', 'size']
dados_numericos = tips[colunas_alvo]

print(f"Assimetria: \n{dados_numericos.skew()}: ")
print(f"\nCurtose: \n{dados_numericos.kurtosis()}")
```

## 5. Correlação

A correlação mede a força e a direção da relação linear entre duas variáveis numéricas. Um coeficiente de correlação próximo de 1 indica uma forte relação linear positiva, -1 indica uma forte relação linear negativa, e 0 indica nenhuma relação linear. A matriz de correlação nos permite visualizar essas relações para todos os pares de variáveis numéricas em nosso dataset.

## 7. Normalização de Dados

A normalização é um processo crucial no pré-processamento de dados, especialmente quando algoritmos de Machine Learning são sensíveis à escala das features (por exemplo, modelos baseados em distância como KMeans, SVMs com kernels, ou Redes Neurais). O objetivo é ajustar os valores das features para uma escala comum, sem distorcer as diferenças nas faixas de valores ou perder informações. Existem duas técnicas comuns:

1.  **Min-Max Scaling (Normalização)**: Escala os dados para um intervalo fixo, geralmente entre 0 e 1. A fórmula é: `X_novo = (X - X_min) / (X_max - X_min)`.
2.  **StandardScaler (Padronização - Z-score)**: Transforma os dados para ter média 0 e desvio padrão 1. A fórmula é: `X_novo = (X - média) / desvio_padrão`.
